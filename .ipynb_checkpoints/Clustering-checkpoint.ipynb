```python
# show the amount of null data in the dataframe
def show_null(df): 
    null_columns = (df.isnull().sum(axis=0)/len(df)).sort_values(ascending = False).index
    
    null_data = pd.concat(
        [
            df.isnull().sum(axis=0),
            (df.isnull().sum(axis=0)/len(df)).sort_values(ascending = False),
            df.loc[:, df.columns.isin(list(null_columns))].dtypes
        ],
        axis=1
    )
    
    null_data = null_data.rename(columns= {0: 'quantity', 1: '% null', 2: 'type'}).sort_values(ascending = False, by = '% null')
    
    return null_data
```

```python
# view numeric data
def plot_num(df, feature):
    fig = plt.figure(figsize=(12,8))
    gs = GridSpec(1,2)
    sns.boxplot(y = feature, data = df, color = 'gold', ax = fig.add_subplot(gs[0,0]))
    plt.ylabel(feature)
    sns.stripplot(y = df[feature], color = 'deepskyblue', ax = fig.add_subplot(gs[0,1]))
    plt.ylabel(None)
    plt.show()
```

```python
# view categorical data
def plot_cat(df, feature):
    fig = plt.figure(figsize=(12,8))
    ax = sns.countplot(data = df, x = feature)
    
    for p in ax.patches:
        ax.annotate(f'\n{p.get_height()}', (p.get_x() + 0.4 , p.get_height() + 5), ha = 'center', color = 'black')
        
    plt.xlabel('\n' + feature)
    plt.ylabel('Number of customers')
    plt.show()
```

## Joining data

```python
df_list = [people, products, promotion, place]
```

```python
final_df = reduce(lambda df1, df2: pd.merge(df1, df2, on='ID'), df_list) # joining all dataframes
```

```python
final_df.shape
```

```python
final_df.columns 
```

```python
final_df.drop(columns=['ID'], inplace=True)
```

# Normalization

```python
def scaler(df): # 0 - 1 normalization
    df2 = df.copy()
    for column in df2:
        #print(f"Column: {column}, dtype: {df2[column].dtype}")
        val_min = df2[column].min()
        val_max = df2[column].max()
        df2[column] = (df[column] - val_min)/(val_max - val_min)
    return df2 
```

```python
scaled_df = scaler(final_df)
```

```python
scaled_df.describe()
```

# Elbow Method

```python
inertia = []
clusters = range(1, 15)

for k in clusters:
    Model = KMeans(n_clusters=k)
    Model.fit(scaled_df) #fit the model
    inertia.append(Model.inertia_)
```

```python
plt.plot(clusters, inertia, marker='o')
plt.xlabel('Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method')
plt.xticks(clusters)
plt.grid(True)
plt.show()
```

# Weight Allocation

```python
scaled_df.columns
```

```python
scaled_df.AllProducts = 8*scaled_df.AllProducts
scaled_df.AllPurchases = 6*scaled_df.AllPurchases
scaled_df.Recency = 4*scaled_df.Recency
scaled_df.NumWebVisitsMonth = 4*scaled_df.NumWebVisitsMonth
scaled_df.Income = 4*scaled_df.Income
scaled_df.LifeTime = 4*scaled_df.LifeTime
scaled_df.Year_Birth = 2*scaled_df.Year_Birth
scaled_df.ChildrenHome = 2*scaled_df.ChildrenHome
scaled_df.Education = 2*scaled_df.Education
```

```python
scaled_df.describe()
```

## Kmeans

```python
Model = KMeans(n_clusters=4)
```

```python
clusters = Model.fit_predict(scaled_df) # fit the model with predict
```

```python
final_df['cluster'] = clusters
```

## Saving the model

```python
# save the model to disk
filename = 'final_model.sav'
pickle.dump(Model, open(filename, 'wb'))
```

Cluster 0 is characterized by making a large volume of purchases, although with a significantly lower average ticket than the previous group. However, even with this lower average ticket, they still contribute significantly to the company's revenue. Customers in this group have a relatively high annual salary, are older, and typically have one child at home. This is the group that shows the greatest appreciation for the wine category and also has the highest number of customers.

Cluster 1 consists of a younger group with lower purchasing power, which is the least active among all clusters. The company needs to be cautious as they are churning and reducing their purchasing behavior compared to before. This group is the least responsive to promotional campaigns. It represents an important opportunity for the company to launch campaigns targeting products aimed at children. They are the group that engages the most in promotional purchases. They prefer shopping in physical stores and are the most frequent visitors to the website. Additionally, they show a preference for gold products compared to other higher-value clusters.

Cluster 2 comprises the youngest customers with lower purchasing power, yet they are the most active within the company. This group has been registered for a shorter period. Customers enjoy shopping during promotions and prefer purchasing in physical stores. They also frequently visit the website, likely to seek out promotions. Additionally, they show a preference for gold products compared to other higher-value clusters.

Cluster 3 is comprised of the highest-value customer group, characterized by a smaller number of members. These customers stand out for their high income and higher level of education. Additionally, they are the biggest spenders, with the highest average ticket among all groups. They do not show a significant preference for visiting the company's website or taking advantage of promotions frequently, opting instead to purchase through catalogs or directly in physical stores. Their preferred product categories include meats and wines. The majority of this group does not have children at home and are loyal customers, registered for a longer period of time.
